```python
import pandas as pd
```

```python
df = pd.read_excel("Group 4 Database.xlsx", header=2)
df.head()
```

```python
df.info()
```

```python
df[df.duplicated()]
```

```python
df.isnull().sum()
```

```python
highest_leave = df.groupby('Department')['Reason for leaving'].count().sort_values(ascending=False)
highest_leave
```

```python
avg_tenure = df.groupby('Department')['Years in organization'].mean().sort_values(ascending=False)
avg_tenure.round(2)
```

```python
top_3_reasons = df['Reason for leaving'].value_counts().head(3)
top_3_reasons
```

```python
most_common_reason = (
    df.groupby(['Department', 'Reason for leaving'])
      .size()
      .reset_index(name='Count')
      .sort_values(['Department', 'Count'], ascending=[True, False])
      .drop_duplicates(subset=['Department'])
)

most_common_reason
```

## Number of employees who left, gender-wise

```python
df.groupby('Gender').size()
```

## Average years in organization by gender

```python
df.groupby('Gender')['Years in organization'].mean().round(2)
```

```python
df['Experience_Group'] = pd.cut(
    df['Years in organization'],
    bins=[0, 3, 7, 100],
    labels=['0–3 Years', '4–7 Years', '8+ Years']
)
```

```python
df['Experience_Group'].value_counts()
```

```python
early_exit = df[df['Years in organization'] <= 3]
early_exit.groupby('Department').size().sort_values(ascending=False)
```
